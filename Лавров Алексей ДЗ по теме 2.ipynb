```python
word = input("Введите слово: ").strip() # Вводим слова и удаляем пробелы из него

length = len(word) # Определяем длину слова

index_middle = length // 2 #Вычисляем индекс середины слова

# Задаем условие
result = word[index_middle-1:index_middle+1] if length % 2 == 0 else word[index_middle]

print(f"Результат: {result}") # Выводим результат
```

```python
# Списки имен
boys = ['Peter', 'Alex', 'John', 'Arthur', 'Richard']
girls = ['Kate', 'Liza', 'Kira', 'Emma', 'Trisha']

#Основной цикл
if len(boys) == len(girls):
    boys.sort()
    girls.sort()
    print('Идеальные пары:')
    for sort_boys, sort_girls in zip(boys, girls):
        print(f'{sort_boys} и {sort_girls}')
else:
    print('Внимание, кто-то может остаться без пары!')
```
